```python
#split the items in each row
df['Select Inverstors Split'] = df['Select Inverstors'].apply( lambda x: x.split(','))
# create a list from the new list
df_list = df['Select Inverstors Split'] 
df_list = df_list.to_frame()
# rename the main column
df_list.rename(columns = {'Select Inverstors Split':'inv_companies'}, inplace = True)
# add each separate item to a new column
df_list[['inv_company_1', 'inv_company_2', 'inv_company_3', 'inv_company_4']] = pd.DataFrame(df_list.inv_companies.tolist(), index= df_list.index)
# add the 'Select Investor' column to the new columns ('inv_company_#')
df_list = df_list.join(df['Select Inverstors'])
# change order of columns to merge with main df
df_list[['Select Inverstors', 'inv_companies', 'inv_company_1', 'inv_company_2', 'inv_company_3', 'inv_company_4']]
# merge to main df
df1 = pd.merge(df, df_list, how = 'left', left_on = 'Select Inverstors', right_on = 'Select Inverstors')
df1
```

# Working with 'Founded Year' columnÂ¶
    1)replace None values to 1900
    2)change STR to datetime

```python
df1['Founded Year'] = df1['Founded Year'].apply(lambda x: x.replace('None','1900')).apply(lambda x: datetime.strptime(x, '%Y'))
```

# Working with Data Joined column
    1)change STR to datetime

```python
df1['Date Joined'] = df1['Date Joined'].apply(lambda x: datetime.strptime(x, '%m/%d/%Y'))
```

```python
unicorns_df = df1[['Company', 'Valuation ($B)', 'Date Joined', 'Country', 'City',
                'Founded Year', 'Total Raised', 'Financial Stage',
                'Investors Count', 'Deal Terms', 'Portfolio Exits', 'Select Inverstors', 'Industry', 'inv_company_1',
                'inv_company_2', 'inv_company_3', 'inv_company_4']]
```

```python
#replacing None for NAN
unicorns_df.replace(to_replace='None', value=np.nan, inplace=True)
```

```python
unicorns_df.drop_duplicates(inplace=True)
```

```python
unicorns_df.to_csv('Updated_Unicorns1.csv')
```
